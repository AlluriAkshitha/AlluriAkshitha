# Fit a Linear Regression

```python
#build regression model

import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily", data = data1).fit()
```

```python
model1.summary()
```

* The predicted value is
* Y (hat) = beta_0 + Beta_1X
* Y (hat) = 13.83556 + 1.3397

- The probability (p-value) for intercept(beta_0) is 0.05
- Thereforethe intercept coefficient may not be that much significant in prediction
- However the p-value for "daily" (beta_1) is 0.00 < 0.05
- Therefore the beta_1 coefficient is highly significant and its contribution to prediction.

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "y", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33

#predicted response vector
y_hat = b0 + b1*x

#plotting the regression line
plt.plot(x, y_hat, color = "b")

#putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#plot the linear regression line using seaborn() method
sns.regplot(x="daily", y="sunday", data=data1)
plt.xlim([0,1250])
plt.show()
```

```python
model1.params
```

```python
#pritn model statistics ( t and p-Values)
print(f'model t-values:\n{model1.tvalues}\n----------\nmodel p-values: \n{model1.pvalues}')
```

## Predict for new data point

```python
#Predict for 200 and 300 daily circulation
newdata = pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
#Predict on all given training data (daily)
pred = model1.predict(data1["daily"])
pred
```

```python
#Add predicted values as a column in data1
data1["Y_hat"] = pred
data1
```

```python
#Compute the error values (residuals) and add as another column
data1["residuals"]= data1["sunday"]-data1["Y_hat"]
data1
```

```python
#compute mean squared error for the model
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ", rmse)
```

### Checking the model for residuals scatter plot 

```python
#Plot the r versus y_hat
plt.scatter(data1["Y_hat"], data1["residuals"])
```

#### Observation
* The residual data points are randomly scattered
* There appears to be no trend and the residuals are randomly placed around the zero error line
* Hence the assumption of homoscedasticty is satisfied ( contant variance in residuals)

```python
#plot Q-Q plot
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```

```python
#plot the kde distribution for residuals
sns.histplot(data1["residuals"], kde=True)
```

#### Observations
* The data points are seen to closely follow the reference line of normality
* Hence the residuals are approximately normally distributed as also can be seen from the kde plot
