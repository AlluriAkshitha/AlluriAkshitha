```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_clean.csv")
data
```

```python
data.info()
```

```python
#DataFrame attributes
print(type(data))
print(data.shape)
print(data.size)
```

```python
#Drop the duplicated column- temp C and Unnamed column
#data1 = data.drop(['Unnamed: 0', 'Temp C'], axis = 1, inplace=True)  #inplace=True to edit the original table permanently
data1 = data.drop(['Unnamed: 0', 'Temp C'], axis = 1)   
data1
```

```python
data1.info()
```

```python
#change the data type of Month object to numeric
data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
#print all duplicated rows
data1[data1.duplicated(keep = False)]
```

```python
data1[data1.duplicated()]
```

```python
data1.drop_duplicates(keep='first',inplace = True)
data1
```

```python
#Renaming of columns
data1.rename({'Solar.R': 'Solar'}, axis=1, inplace = True)
data1
```

Impute the missing values

```python
#Display data1 missing values count in each column using isnull().sum()
data1.isnull().sum()
```

```python
#Visualize the data1 missing values using graph

cols = data1.columns
colors = ['pink', 'purple']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
#Find the mean and median values of each numeric column
#Imputation of missing value with median
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
#Replace the ozone missing values with median value
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
median_Solar = data1["Solar"].median()
mean_Solar = data1["Solar"].mean()
print("Median of solar: ", median_Solar)
print("Mean of solar: ", mean_Solar)
```

```python
data1['Solar'] = data1['Solar'].fillna(mean_Solar)
data1.isnull().sum()
```

```python
data1.head()
```

```python
#mode values of column weather
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
#Impute missing values-Replace NaN with mode 
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
print(data1["Month"].value_counts())
mode_month = data1["Month"].mode()[0]
print(mode_month)
```

```python
data1["Month"] = data1["Month"].fillna(mode_month)
data1.isnull().sum()
```
